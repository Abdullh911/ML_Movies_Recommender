```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv('D:\ml practice\TMDB_movie_dataset_v11.csv')
df['release_date'] = pd.to_datetime(df['release_date'])
df = df[(df['vote_average'] != 0) & (df['release_date'].dt.year >= 2000)]
df.head()
```

```python
df_copy=df.copy()
df_copy.head()
```

```python
df['combined_features'] = df['genres'] + ' ' + df['keywords'] + ' ' + df['tagline'] + ' ' + df['production_countries']
df.head()
```

```python
df = df.dropna(subset=['combined_features'])
df['index'] = range(len(df))
df.head()
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['combined_features'])
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim
```

```python
import difflib
movie_name = input(' Enter your favourite movie name : ')
list_of_all_titles = df['title'].tolist()
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
close_match = find_close_match[0]
close_match
```

```python
index_of_the_movie = df[df.title == close_match]['index'].values[0]
print(df.iloc[index_of_the_movie,3])
similarity_score=list(enumerate(cosine_sim[index_of_the_movie]))
```

```python
similarity_score=sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python

for i in range(50):
    ind = similarity_score[i][0]
    movie_name = df.iloc[ind, 1] 
    overview = df.iloc[ind, 15]  
    rate = df.iloc[ind, 2]        
    print('Movie : ' + movie_name)
    print('Overview: ' + overview)
    print('Rate: ' + str(rate))
    print('----------------------------')
```
